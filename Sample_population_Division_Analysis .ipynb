# A pie chart is a type of data visualization.
# while Matplotlib is a plotting library in Python that can be used to create various types of visualizations, including pie charts.

```python
import matplotlib.pyplot as plt

# Data for the pie chart
labels = ['Christians', 'Muslims', 'Hindus', 'Buddhists', 'Jews', 'Folk Religions', 'Other Religions', 'Unaffiliated']
sizes = [32, 23, 15, 7, 0.2, 6, 1, 16]  # Percentages
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6','#c2f0c2','#ffccff']
explode = (0.1, 0, 0, 0, 0, 0, 0, 0)  # explode the 1st slice (Christianity)

# Create the pie chart
plt.figure(figsize=(10, 7))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Global Religious Demographics (Percentage)')
plt.show()
```

# Stacked bar charts are better for showing total values and their components, while ribbon graphs are more effective for illustrating relationships and changes over time.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the ribbon chart
regions = ['Asia-Pacific', 'Middle East & North Africa', 'Sub-Saharan Africa']
religions = ['Hindus', 'Buddhists', 'Muslims', 'Unaffiliated']

# Data values
data = {
    'Hindus': [99, 0, 0],
    'Buddhists': [99, 0, 0],
    'Muslims': [62, 20, 16],
    'Unaffiliated': [76, 0, 0]
}

# Create a stacked bar chart
bar_width = 0.35
index = np.arange(len(regions))

# Create bars for each religion
bottom_values = np.zeros(len(regions))  # Initialize bottom values for stacking
for religion in religions:
    plt.bar(index, data[religion], bar_width, label=religion, bottom=bottom_values)
    bottom_values += np.array(data[religion])  # Update bottom values for the next stack

# Add labels and title
plt.xlabel('Regions')
plt.ylabel('Percentage (%)')
plt.title('Geographic Distribution of Major Religions')
plt.xticks(index, regions)
plt.legend()
plt.ylim(0, 100)
plt.grid(axis='y')

# Show the plot
plt.tight_layout()
plt.show()
```

# A waterfall chart visually represents the cumulative effect of sequentially introduced positive or negative values

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the waterfall chart
regions = ['Asia-Pacific', 'Middle East & North Africa', 'Sub-Saharan Africa']
religions = ['Hindus', 'Buddhists', 'Muslims', 'Unaffiliated']

# Data values
data = {
    'Hindus': [99, 0, 0],
    'Buddhists': [99, 0, 0],
    'Muslims': [62, 20, 16],
    'Unaffiliated': [76, 0, 0]
}

# Prepare the waterfall data
waterfall_data = np.zeros((len(regions), len(religions) + 1))

# Fill the waterfall data
for i, religion in enumerate(religions):
    waterfall_data[:, i + 1] = waterfall_data[:, i] + np.array(data[religion])

# Create the waterfall chart
fig, ax = plt.subplots(figsize=(10, 6))

# Create bars for each religion
for i, religion in enumerate(religions):
    ax.bar(regions, data[religion], bottom=waterfall_data[:, i], label=religion)

# Add labels and title
ax.set_xlabel('Regions')
ax.set_ylabel('Percentage (%)')
ax.set_title('Geographic Distribution of Major Religions')
ax.set_ylim(0, 100)
ax.legend()
ax.grid(axis='y')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic
import pandas as pd

# Data preparation
data = {
    'Region': ['North America', 'Europe', 'Asia-Pacific', 'Middle East & North Africa', 'Sub-Saharan Africa'],
    'Unaffiliated Population (millions)': [280, 180, 760, 10, 20]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a mosaic plot
plt.figure(figsize=(10, 6))
mosaic(df, ['Region'], title='Distribution of Religiously Unaffiliated Population by Region',
       labelizer=lambda k: f"{k[0]}: {df.loc[df['Region'] == k[0], 'Unaffiliated Population (millions)'].values[0]}M")
plt.ylabel('Population (millions)')
plt.show()
```

A cumulative percentage is a running total of percentages that shows the accumulation of values over a dataset, allowing for the identification of the proportion of the total that each category contributes as you progress.

helping to identify the most significant factors in a dataset based on the Pareto principle (80/20 rule), which states that roughly 80% of effects come from 20% of causes.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Hypothetical data
data = {
    'Religious Group': ['Christians', 'Muslims', 'Hindus', 'Buddhists', 'Jews', 'Folk Religions', 'Unaffiliated'],
    'Population (millions)': [2200, 1600, 1000, 500, 14, 400, 1100]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Sort the DataFrame by Population in descending order
df = df.sort_values(by='Population (millions)', ascending=False)

# Calculate cumulative percentage
df['Cumulative Percentage'] = df['Population (millions)'].cumsum() / df['Population (millions)'].sum() * 100

# Create the Pareto chart
fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar chart for the population
ax1.bar(df['Religious Group'], df['Population (millions)'], color='skyblue')
ax1.set_xlabel('Religious Group')
ax1.set_ylabel('Population (millions)', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')
ax1.set_title('Pareto Chart of Religious Groups by Population')

# Create a second y-axis for the cumulative percentage
ax2 = ax1.twinx()
ax2.plot(df['Religious Group'], df['Cumulative Percentage'], color='orange', marker='o', linewidth=2)
ax2.set_ylabel('Cumulative Percentage (%)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Show grid
ax1.grid(axis='y')

# Show the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the pie chart
labels = ['Christians', 'Muslims']
sizes = [2300, 1800]  # Population in millions
colors = ['#ff9999', '#66b3ff']
explode = (0.1, 0)

# Pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Global Population Distribution of Christians and Muslims (in millions)')
plt.show()

# Data for the line chart
years = [2000, 2005, 2010, 2015, 2020, 2025]
christian_population = [2100, 2200, 2300, 2350, 2400, 2450]  # Projected population in millions
muslim_population = [1400, 1500, 1600, 1700, 1800, 1900]  # Projected population in millions

# Line chart
plt.figure(figsize=(10, 6))
plt.plot(years, christian_population, marker='o', label='Christians', color='blue')
plt.plot(years, muslim_population, marker='o', label='Muslims', color='orange')
plt.title('Projected Population Growth of Christians and Muslims')
plt.xlabel('Year')
plt.ylabel('Population (in millions)')
plt.xticks(years)
plt.legend()
plt.grid()
plt.show()
```
